```python
#residence geography= no data for workplace geog
import requests, pandas as pd

year='2019'
dsource='acs'
dname='acs5'
res='B08301_001E,B08301_002E,B08301_003E,B08301_004E,B08301_010E,B08301_016E,B08301_017E,B08301_018E,B08301_019E,B08301_020E,B08301_021E'
#work='B08601_001E,B08601_002E,B08601_003E,B08601_004E,B08601_010E,B08601_016E,B08601_017E,B08601_018E,B08601_019E,B08601_020E,B08601_021E'

state='15'
county='003' #hawaii=001, oahu=003, maui=009

#Base url
base_url = f'https://api.census.gov/data/{year}/{dsource}/{dname}'

data_url= f'{base_url}?get={res}&for=tract:*&in=state:{state}&in=county:{county}'

#data_url = f'{base_url}?get={cols1}&for=county:{county}&in=state:{state}'
response=requests.get(data_url)

#print(response.text)

#format df
data=response.json()
df=pd.DataFrame(data[1:], columns=data[0]).\
    rename(columns={"B08301_001E":"Tot", 
                    "B08301_002E":"tot_car", 
                    "B08301_003E":"da",
                    "B08301_004E":"cp",
                    "B08301_010E":"pt",
                    "B08301_016E":"tx",
                    "B08301_017E":"mc",
                    "B08301_018E":"bk",
                    "B08301_019E":"wk",
                    "B08301_020E":"other",
                    "B08301_021E":"home",
                    "block group":"bg"})
df['GEOID']=df.state+df.county+df.tract
df.drop(columns=['state','county','tract'],inplace=True)
df_res19=df.copy()
```

```python
df_res19=df.copy()
df_res19 = df_res19.astype('int64')
df_res19['GEOID'] = df_res19['GEOID'].astype('str')

df_res19['TT'] = df_res19['da'] + df_res19['cp'] + df_res19['pt'] + df_res19['wk'] + df_res19['home']
df_res19 = df_res19[['GEOID','da','cp','pt','wk','home','TT']]
df_res19.head()
```

```python
#!pip install geopandas
import geopandas as gpd
zipfile = "https://www2.census.gov/geo/tiger/GENZ2018/shp/cb_2018_15_tract_500k.zip"
dmap = gpd.read_file(zipfile)
dmap=dmap.loc[(dmap['COUNTYFP'] == '003') & (dmap['GEOID'] != '150039812001')]
dmap=dmap.loc[:, ['GEOID','geometry']]
dmap.crs = "EPSG:4326"
```

```python
#df_clean['GEOID'] = df_clean['GEOID'].astype(str).replace('\.0', '', regex=True)
df_join = dmap.merge(df_res19, on='GEOID')
df_join['geoid'] = df_join.index.astype(str)
```

```python
df_join.head()
```

```python
import folium
m = folium.Map(location=[21.461595, -157.954415], zoom_start=11, tiles = 'cartodbpositron')

c=folium.Choropleth(
    geo_data=df_join,
    data=df_join,
    key_on='feature.id',
    columns=['geoid','TT'],
    fill_color="YlOrRd",
    fill_opacity=0.7,
    line_opacity=0.9,
    line_color='white', 
    line_weight=0,
    highlight=False, 
    smooth_factor=1.0,
    legend_name="Population").add_to(m)

for key in c._children:
    if key.startswith('color_map'):
        del(c._children[key])
        
c.add_to(m)
m
```
